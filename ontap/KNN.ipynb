```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, recall_score, precision_score, classification_report
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Câu 3
```

```python

# Load dữ liệu wine
wine = load_wine()
X = wine.data
y = wine.target
```

```python
# Chia 70:30 (theo yêu cầu thầy)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# tạo ra và train KNN với k=5
KNN_new = KNeighborsClassifier(n_neighbors=5)
KNN_new.fit(X_train, y_train)

# kết quả dự đoán
y_pred = KNN_new.predict(X_test)
```

```python
# Với k = 5 thì cho ra được kết quả
print("Kết quả với k=5:")
print(f"Độ chính xác (Accuracy): {accuracy_score(y_test, y_pred):.4f}")
print(f"Độ nhạy trung bình (Recall): {recall_score(y_test, y_pred, average='macro'):.4f}")
print(f"Độ chính xác trung bình (Precision): {precision_score(y_test, y_pred, average='macro'):.4f}")
```

```python
print("\n report:")
print(classification_report(y_test, y_pred, target_names=wine.target_names))
```

```python
# test k khác nhau
k_values = [1, 3, 5, 7, 9]
accuracies = []
recalls = []
precisions = []
```

```python
for k in k_values:
    KNN = KNeighborsClassifier(n_neighbors=k)
    KNN.fit(X_train, y_train)
    y_pred = KNN.predict(X_test)
    
    accuracies.append(accuracy_score(y_test, y_pred))
    recalls.append(recall_score(y_test, y_pred, average='macro'))
    precisions.append(precision_score(y_test, y_pred, average='macro'))
```

```python
# Vẽ đồ thị
plt.figure(figsize=(12, 6))

plt.plot(k_values, accuracies, marker='o', label='Accuracy', linewidth=2)
plt.plot(k_values, recalls, marker='s', label='Recall', linewidth=2)
plt.plot(k_values, precisions, marker='^', label='Precision', linewidth=2)

plt.xlabel('Số lượng KNN (k)')
plt.ylabel('Điểm số')
plt.title('Mô hình KNN với value k khác nhau')
plt.grid(True)
plt.legend()
plt.xticks(k_values)
plt.show()
```

```python
# Report 
print("\n Report value k:")
print("\nk\tAccuracy\tRecall\t\tPrecision")
print("-" * 50)
for k, acc, rec, prec in zip(k_values, accuracies, recalls, precisions):
    print(f"{k}\t{acc:.4f}\t\t{rec:.4f}\t\t{prec:.4f}")
```
